## Pandas CSV reader & basic analysis
- Read CSV/Excel into DataFrame and inspect head/tail/types.
- Compute summary stats (mean, median, min, max, count).
- Filter rows, select columns and slice subsets.
- Save filtered results to CSV/excel.

```python
import pandas as pd
```

### Read CSV / Excel File

```python
# Read CSV
data = pd.read_csv("genz_college_admission_prediction.csv")
```

### Inspect Data

```python
print("First 5 Rows:")
data.head()
```

```python
print("Last 5 Rows:")
data.tail()
```

```python
print("Data Types:")
print(data.dtypes)
```

```python
print("Dataset Shape:")
print(data.shape)
```

### Summary Statistics

```python
print("Summary Statistics")
numeric_cols = data.select_dtypes(include="number")

print("Mean:")
print(numeric_cols.mean())
```

```python
print("Median:")
print(numeric_cols.median())
```

```python
print("Minimum Values:")
print(numeric_cols.min())
```

```python
print("Maximum Values:")
print(numeric_cols.max())
```

```python
print("Count:")
print(numeric_cols.count())
```

### Select Column

```python
selected_columns = data[['admission_status', 'age']]  

print("Selected Columns:")
print(selected_columns.head())
```

### Filter Rows

```python
filtered_rows = data[data['age'] > 18]  

print("\nFiltered Rows (age > 18):")
filtered_rows.head()
```

### Slice Subsets

```python
subset = data.iloc[0:10, 0:3]

print("Subset (First 10 Rows, First 3 Columns):")
print(subset)
```

### Save Filtered Results

```python
filtered_rows.to_csv("filtered_data.csv", index=False)
print("Filtered data saved successfully!")
```
