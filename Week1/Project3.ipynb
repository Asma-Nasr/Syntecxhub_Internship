## Data cleaning utility
- Detect and handle missing values (drop / fill /impute).
- Fix incorrect dtypes (dates, numbers) and parse dates.
- Remove duplicates and standardize column names.
- Output a cleaned dataset and brief cleaning log.

```python
import pandas as pd
import numpy as np
```

```python
# Load dataset
df = pd.read_csv("genz_college_admission_prediction.csv")

# Store cleaning log
cleaning_log = []
```

```python
df.head(3)
```

### Handle Missing Values

```python
df.isna().sum().sum() # no missing values
```

### Fix Data Types

```python
df.info()
```

```python
# convert object columns to category
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].astype("category")
cleaning_log.append("Converted object columns to category data type")
```

### Remove Duplicates

```python
df.duplicated().sum() # no duplicates
```

### Final Dataset Info

```python
cleaning_log.append(
    f"Final dataset shape: {df.shape[0]} rows, {df.shape[1]} columns."
)
```

### Save Cleaned Dataset

```python
df.to_csv("cleaned_dataset.csv", index=False)
with open("cleaning_log.txt", "w") as file:
    file.write("DATA CLEANING LOG\n")
    file.write("=" * 30 + "\n")

    for entry in cleaning_log:
        file.write(entry + "\n")
```

### Display Log

```python
print("Data Cleaning Finished\n")
for entry in cleaning_log:
    print("-", entry)

print("Cleaned dataset")
print("Cleaning log saved")
```
