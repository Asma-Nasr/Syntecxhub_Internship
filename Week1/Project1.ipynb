## NumPy Data Explorer
- Learned NumPy fundamentals – array creation, indexing, and slicing.
- Performed mathematical, axis-wise, and statistical operations on datasets.
- Applied reshaping and broadcasting techniques for efficient computation.
- Implemented save/load operations for NumPy arrays.
- Compared NumPy’s performance with standard Python lists.

```python
import numpy as np
import time
```

### Array creation

```python
data = np.array([
    [41, 84, 92],
    [25, 88, 88],
    [34, 72, 72],
    [12, 91, 81]])
print(data)
```

### Indexing and Slicing

```python
print("First Row:")
print(data[0])
```

```python
print("Second Column:")
print(data[:, 1])
```

```python
print("First Two Rows:")
print(data[:2])
```

###  Mathematical Operations

```python
print("Dataset + 10:")
print(data + 10)
```

```python
print("Dataset Squared:")
print(data ** 2)
```

### AXIS-WISE OPERATIONS

```python
print("Row-wise Sum:")
print(np.sum(data, axis=1))
```

```python
print("Column-wise Sum:")
print(np.sum(data, axis=0))
```

### Statistical Operations

```python
print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Maximum:", np.max(data))
print("Minimum:", np.min(data))
print("Standard Deviation:", np.std(data))
```

###  Reshaping

```python
reshaped = data.reshape(2, 6)
print("Reshaped Array (2x6):")
print(reshaped)
```

### Broadcasting

```python
bonus_marks = np.array([5, 3, 2])
updated_scores = data + bonus_marks
print("After Broadcasting Bonus Marks:")
print(updated_scores)
```

### Save and Load Arrays

```python
np.save("student_scores.npy", data)
loaded_data = np.load("student_scores.npy")
print("\nLoaded Data:")
print(loaded_data)
```

### Numpy vs Python list Performance

```python
size = 1_000_000
# Python List
py_list = list(range(size))
start = time.time()
py_result = [x * 2 for x in py_list]
python_time = time.time() - start
# NumPy Array
np_array = np.arange(size)
start = time.time()
np_result = np_array * 2
numpy_time = time.time() - start

print("Performance Comparison")
print(f"Python List Time : {python_time:.6f} seconds")
print(f"NumPy Array Time : {numpy_time:.6f} seconds")

if numpy_time < python_time:
    print("NumPy is faster.")
else:
    print("Python list is faster.")
```
